## Генератор данных (2 балла)

За вас реализован шаблонный скрипт по созданию CSV-файла с данными. Ваша задача — осознать его и подстроить под свою тематику: придумать названия колонок и добавить подходящие генераторы значений. Тематика любая — звери, люди, товары в магазине, что угодно.

После этого прогоните скрипт локально и убедитесь, что он работает и создаёт корректный CSV-файл.

Основная задача — упаковать этот скрипт в Docker-контейнер, который после запуска формирует файл `data.csv` в директории `data/` **локально на хосте**. Подсказка: вам предстоит разобраться, что такое монтирование томов, вспомнить как правильно структурировать необходимые файлы, чтобы собрать контейнер

**Ключевой чекпоинт** — через `run.sh` должны работать две команды:

```bash
./run.sh build_generator   # собрать образ для контейнера генератора
./run.sh run_generator     # запустить контейнер, который сгенерирует data/data.csv локально
./run.sh create_local_data # в директории local_data создает data.csv (для локальной отладки)
```

> **Запрещено:** вносить изменения в `generate.py`, кроме названий колонок и количества генерируемых строк, даже когда будете прогонять скрипт локально
